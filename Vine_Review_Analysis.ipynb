```python
import pandas as pd
```

```python
#read in csv as data frame 
vine_df = pd.read_csv("vine_table.csv")
vine_df.head()
```

```python
#filter total votes columns to products that have greater than or equal 20 total votes
better_vine_table_df = vine_df[vine_df["total_votes"] >=20]
better_vine_table_df
```

```python
#helpful votes divided by total votes is greater than or equal to 50 percent 
better_helpful_votes_df = better_vine_table_df[better_vine_table_df["helpful_votes"] / better_vine_table_df["total_votes"] >= .5]
better_helpful_votes_df
```

```python
#paid reviews
paid_vine_df = better_helpful_votes_df[better_helpful_votes_df["vine"] == "Y"]
paid_vine_df
```

```python
#unpaid reviews
unpaid_vine_df = better_helpful_votes_df[better_helpful_votes_df["vine"] == "N"]
unpaid_vine_df
```

```python
#total reviews
total_reviews = better_helpful_votes_df["review_id"].count()
total_reviews
```

```python
#total 5 star reviews
five_star_reviews_df = better_helpful_votes_df[better_helpful_votes_df["star_rating"] == 5.0]
total_five_star_reviews = five_star_reviews_df["star_rating"].count()
total_five_star_reviews
```

```python
#percentage of 5 star paid reviews
five_star_reviews_paid = paid_vine_df[paid_vine_df["star_rating"] == 5.0]["star_rating"].count()
per_five_star_reviews_paid = (five_star_reviews_paid / total_five_star_reviews)*100
per_five_star_reviews_paid
```

```python
#percentage of 5 star unpaid reviews
five_star_reviews_unpaid = unpaid_vine_df[unpaid_vine_df["star_rating"] == 5.0]["star_rating"].count()
per_five_star_reviews_unpaid = (five_star_reviews_unpaid / total_five_star_reviews)*100
per_five_star_reviews_unpaid
```
